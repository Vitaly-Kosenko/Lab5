<h2> Функція № 1 спискові включення. Генерація списків цілочисельного ділення на 7 з використанням лямбда-функції</h2>

```python
s = [7, 22, 49, 701]  # Вихідний список чисел
n = list(map(lambda x: x // 7, s))  # Новий список цілочисельного ділення на 7
print(n)  # [1, 3, 7, 100]
```

<h2> Функція № 2 спискові включення. Генерація списку залишків від ділення чисел на 5 </h2>

```python
def mod5(x):      # Створюємо функцію, яка повертає
    return x % 5  # залишок від ділення на 5
s = [5, 7, 11, 20]  # вихідний список
print(list(map(mod5, s)))  # [0, 2, 1, 0] - список залишків
```

<h2> Функція № 3 спискові включення. Генерація списку кубів для чисел від 0 до 9 з використанням функції range()</h2>

```python
s = []  # Створюємо порожній список
for i in range(10):  # Здійснюємо 10 ітерацій - від 0 до 9
    s.append(i ** 3)  # Додаємо до списку куб кожного числа
print(s)  # [0, 1, 8, 27, 64, 125, 216, 343, 512, 729]
```

<h2> Функція № 4 спискові включення. Генерація списку заголовних букв з слова «Парашут»</h2>

```python
# Так як рядок є ітеріруемим об'єктом, то по ній можна пройтися в циклі.
word = "Парашут"  # початкове слово
s = []  # Створюємо порожній список
for i in word.upper():  # Проходимся по кожній букві в слові
    s.append(i)  # Наводимо всі букви до верхнього регістру
print(s)  # ['П', 'А', 'Р', 'А', 'Ш', 'У', 'Т']
```
